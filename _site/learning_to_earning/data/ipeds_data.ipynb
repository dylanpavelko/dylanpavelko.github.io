```python
completions = pd.read_csv('ipeds/completions/c2021_a.csv')
completions
```

```python
frames = [completions, directory_info]
result = pd.merge(completions, directory_info, on="UNITID")
result
```

```python
simplified_result=result[['UNITID','CIPCODE','CTOTALT','INSTNM','STABBR','AWLEVEL']].copy()
simplified_result
```

```python
charges = pd.read_csv('ipeds/charges_by_program/ic2021_ay.csv')
charges[['UNITID','TUITION1','FEE1']]
```

```python
result = pd.merge(simplified_result, charges[['UNITID','CHG2AY3']], on="UNITID")
result
```
